# Placing WTWY Street Teams to Align with Gendered Trip-Chaining Behavior 


The goal of this project is to effectively place street teams from WomenTechWomenYes (WTWY) at NYC subway stations in order to maximize reach to folks who may be interested in and passionate about their cause. I plan to analyze the stations' proximity to schools and daycares in order to take advantage of women's 'trip-chaining' behaviors, that is, accomplishing multiple errands at once including picking up and dropping off children, according to the 2019 book *Invisible Women*.   

To begin exploring the data, I took nine months of turnstile data from Jan 2021-Sept 2021, and began to filter it down to the most relevant.
